```python
def removeZeros(num):
    clean=""
    for i in range(len(num)):
        if num[i]!="0":
            clean+=num[i:]
            return clean
    return "0"
        
        
```

```python
def positive(num1,num2):
    bigDigit=max(len(num1),len(num2))
    if len(num1)<len(num2):
            small=num1
            large=num2
    else:
        small=num2
        large=num1
    difference=len(large)-len(small)
    carry=0
    answer=""
    sum=0
    for i in range(bigDigit-1,-1,-1):
        
        if i-difference<0:
            a=large[i]
            b="0"
            
        else:
            a=large[i]
            b=small[i-difference]
        sum=int(a)+int(b)+carry
        carry=sum//10
        sum%=10
        
        answer=str(sum)+answer
    if carry!=0:
        answer=str(carry)+answer
    return removeZeros(answer)




```

```python
def negative(larger,smaller):
    difference=len(larger)-len(smaller)
    answer=""
    sign=""
    
    if larger[0]=='-':
        sign="-"
    for i in range(len(larger)-1,-1,-1):
        
        if i-difference<0 or smaller[i-difference]=="-":
            b=0
        else:
            b=int(smaller[i-difference])
        if larger[i]=="-":
            break
        elif larger[i]=="(":
            a=int(larger[i+1:i+3])
            larger=larger[:i]
        else:
            a=int(larger[i])
        
        if a<b:
            a=10+a
            if int(larger[i-1])==0:
                    midValue="("+str(int(larger[i-1])-1)+")" #puts the -1 inside a bracket so that the loop doesn't break
            else:
                midValue=str(int(larger[i-1])-1)
            
            larger=larger[:i-1]+midValue+larger[i:]
        
        answer=removeZeros(str(a-b)+answer)
    answer=sign+answer
    
    return answer
        
            
```

```python
def add(num1,num2):
    
    if num1[0]=="-" and num2[0]=="-":
        return "-"+positive(num1[1:],num2[1:])
    elif num1[0]=="-" or num2[0]=="-":
        larger=max(int(num1),int(num2))
        absLarger=max(abs(int(num1)),abs(int(num2)))
        absMin=min(abs(int(num1)),abs(int(num2)))
        
        if larger==absLarger:
            return negative(str(larger),str(absMin))
        else:
            return negative("-"+str(absLarger),str(larger))
        
        
    else:
        return positive(num1,num2)
    

    
    
        
```

```python
def mult2(num1,num2):
    sign=""
    
    if num1[0]=='-' and num2[0]=="-":
        sign=''
        num1=num1[1:]
        num2=num2[1:]
        
    elif num1[0]=='-':
        sign="-"
        num1=num1[1:]
    elif num2[0]=='-':
        sign='-'
        num2=num2[1:]
    
    if num1=="0" or num2=="0":
        return "0"
    
    
    ans="0"
    count=0
    results={}
    for i in range(len(num1)-1,-1,-1):
        carry=0
        temp=""
        if i not in results:
            for j in range(len(num2)-1,-1,-1):
                multResult=int(num1[i])*int(num2[j])+carry
                carry=multResult//10
                multResult%=10
                temp=str(multResult)+temp
            if carry!=0:
                temp=str(carry)+temp
            results[i]=temp
        else:
            temp=results[i]
        temp=temp+"0"*(len(num1)-1-i)
        
        ans=add(temp,ans)
    return sign+ans

            
            
```

```python
def checkEligible(num1,num2):
    if len(num1)>len(num2):
        return True
    elif len(num1)<len(num2):
        return False
    
    else:
        for i in range(len(num1)):
            
            if int(num1[i]) < int(num2[i]):
                return False
            elif int(num1[i]) > int(num2[i]):
                return True
                                    
    return True
    
checkEligible("110","102")  
```
