```python
class Stack:
    
    def __init__(self):
        self.list=[]
       
        
    def push(self,item):
        self.list.append(item)
        
    def length(self):
        return len(self.list)
    
    def top(self):
        if self.list!=[]:
            return self.list[len(self.list)-1]
        return None
    def pop(self):
        if self.list!=[]:
            return self.list.pop()
        
        return None
```

```python
class Queue:
    def __init__(self):
        self.list=[]
    
    def push (self,item):
        self.list.append(item)
    def pop(self):
        if self.list!=[]:
            return self.list.pop(0)
        return None
    def length(self):
        return len(self.list)
```

```python
def preFix(statement):
    items=Stack()
    numbers=Stack()
    statement=statement.split(" ")
    
    for i in statement:
        items.push(i)
            
    
    while items.length()!=0:
        
        operator=items.pop()
        while operator!="/" and  operator!="*" and operator!="+" and operator!="-":
            numbers.push(operator)
            operator=items.pop()
        
        y=int(numbers.pop())
        x=int(numbers.pop())
        
        
        if operator=="/":
            numbers.push(y/x)
        elif operator=="*":
            numbers.push(y*x)
        elif operator=="+":
            numbers.push(y+x)
        elif operator=="-":
            numbers.push(y-x)
        
        
    return numbers.pop()
print(preFix("+ 4 * 3 12"))
print(preFix("- + * 4 5 3 10"))        
    
```

```python
def preFixBetter(statement):
    items=Stack()
    statement=statement.split(" ")
    for i in statement:
        items.push(i)
        
    
    while items.length()!=1:
        
        operator=items.pop()
        operands=[]
        while operator!="/" and  operator!="*" and operator!="+" and operator!="-":
            operands.append(operator)
            operator=items.pop()
        if len(operands)==1:
            return operands[0]
        x=int(operands[0])
        y=int(operands[1])
        
        if operator=="/":
            items.push(y/x)
        elif operator=="*":
            items.push(y*x)
        elif operator=="+":
            items.push(y+x)
        elif operator=="-":
            items.push(y-x)
    
    return items.pop()       
print(preFixBetter("+ 4 * 3 12"))
print(preFixBetter("- + * 4 5 3 10"))
```

```python
def evenBrackets(state):
    openBracket=0
    squareBracket=0
    curlyBracket=0
    
    for i in state:
        
        if i=='[':
            openBracket+=1
        elif i==']':
            if openBracket ==0:
                return False
            openBracket-=1
        elif i=='[':
            squareBracket+=1
        elif i==']':
            if squareBracket==0:
                return False
            squareBracket-=1
        elif i=='{':
            curlyBracket+=1
        elif i=='}':
            if curlyBracket==0:
                return False
            curlyBracket-=1
    
    return not bool(openBracket+squareBracket+curlyBracket)
            
      
evenBrackets("{(})")       
            
```

```python
def evenBracketsCorrect(state):
    items=Stack()
    
    for i in state:
        if i=='[':
            items.push(i)
        elif i=='(':
            items.push(i)
        elif i=='{':
            items.push(i)
        
        elif i==']':
            if items.pop()!='[':
                return False
        elif i=='}':
            if items.pop()!='{':
                return False
        elif i==')':
            if items.pop()!='(':
                return False
    if items.length()!=0:
        return False
    
    return True
        
evenBracketsCorrect("()")      
```

```python
def rearrangeCodes(barcodes):
    mainStack=Stack()
    tempStack=Stack()
    
   
    counter =[0 for _ in range(max(barcodes)+1)]
    mapped={}
    answer=[]
    
    for i in barcodes:   # holds the frequency of each element 
        counter[i]+=1
    
    for j in range(len(counter)):
        if (counter[j]!=0):
            if counter[j] not in mapped:
                mapped[counter[j]]=[]
            mapped[counter[j]].append(j)
    
    counter=sorted(counter,reverse=True)
    
    for i in counter:
        
        if i not in mapped:
            continue
        
        for j in mapped[i]:
            answer+=[j for k in range(i)]
        del mapped[i]
    
    barcodes=answer
    
    
    
    for i in barcodes:
        if i!=mainStack.top():
            mainStack.push(i)
            
            if tempStack.top()!=None and i!=tempStack.top():
                mainStack.push(tempStack.pop())
        else:
            tempStack.push(i)
    
    return mainStack.list
rearrangeCodes([1,1,1,2,2,2])
```

```python
def countSort(lst):
    counter =[0 for _ in range(max(lst)+1)]
    answer= [0 for _ in lst]
    for i in lst:   # holds the frequency of each element 
        counter[i]+=1
    
    for j in range(1,len(counter)):
        counter[j]+=counter[j-1]
    
    
    for k in range(len(lst)-1,-1,-1):
        answer[counter[lst[k]]-1]=lst[k]
        counter[lst[k]]-=1
    
    return answer

        
```
